# groupby function

```python
#applying groupby and aggregte fucntion in one go.
print(df.groupby(['SepalLengthCm','SepalWidthCm'])['PetalLengthCm'].agg(["max", "min", "sum", "count"]))
```

```python
print(df.groupby(["Species",'SepalLengthCm'])['PetalLengthCm'].agg(["max", "min", "sum", "count"]))
```
