```python
#7 Print Inner Join , Outer Join , Left Join , Right Join by using CITY column
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando" ,"baltimore"],
    "temperature": [21,14,35,32],
    "Population" :[1 , 2 , 3 , 4]
})

df2 = pd.DataFrame({
    "city": ["chicago","new york","San fransisco"],
    "humidity": [65,68,75],
    "Rain" :[15,20,30],
    "country" :["a" , "b" , "c"]
})
```

```python
# inner_join

pd.merge(df1,df2, on = 'city' , how = "inner")
```

```python
# outer_join

pd.merge(df1,df2, on = 'city' , how = "outer")
```

```python
# left_join

pd.merge(df1,df2, on = 'city' , how = "left")
```

```python
# Right join

pd.merge(df1,df2, on = 'city' , how = "right")
```
